```python
df_sorted_oldest = df_single_visitor_per_client.sort_values(by=['client_id', 'date_time'], ascending=True)

df_oldest_step_per_client = df_sorted_oldest.drop_duplicates(subset='client_id', keep='first')

clients_to_remove = df_oldest_step_per_client[df_oldest_step_per_client['process_step'] != 'confirm']['client_id']

df_final_filtered = df_single_visitor_per_client[~df_single_visitor_per_client['client_id'].isin(clients_to_remove)]
```

```python
df_full = pd.merge(df_merged_completion, df_client[['client_id', 'clnt_age']], on='client_id', how='inner')
```

```python
df_full['age_group'] = pd.cut(df_full['clnt_age'], bins=[0, 25, 35, 45, 55, 65, 75, 85, 95, 105], labels=["0-25", "26-35", "36-45", "46-55", "56-65", "66-75", "76-85", "86-95", "96-105"])
completion_rate_by_age = df_full.groupby(['age_group', 'Variation'])['is_completion'].mean().unstack()
```

```python
completion_rate_by_age.plot(kind='bar', figsize=(14, 7))
plt.title('Completion Rate by Age Group and Variation')
plt.xlabel('Age Group')
plt.ylabel('Completion Rate')
plt.xticks(rotation=45)
plt.legend(title='Variation')
plt.tight_layout()
plt.show()
```
